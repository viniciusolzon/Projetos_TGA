```python
class Grafo:
  def __init__(self, txt_file):
    with open(txt_file) as f:
      self.num_vertices = int(f.readline().rstrip())
      lines = f.read()
      lines = lines.split('\n')

    self.matrizAdj = []
    self.listaAdj = []
    i = 1
    for line in lines:
      row = []
      adjacent = [i]
      j = 1
      for vertice in line.split(' '):
        if vertice != '':
          row.append(int(vertice))
          if int(vertice) == 1:
            adjacent.append(int(j))
          j+=1
      self.matrizAdj.append(row)
      self.listaAdj.append(adjacent)
      i+=1

    self.matrizAdj.pop(-1)
    self.listaAdj.pop(-1)

    # para profundidade
    self.profundidade_entrada = []
    self.profundidade_saida = []
    self.pai = []
    self.tempo = 0
    self.matrizCores = [['branco' for i in range(self.num_vertices) ] for j in range(self.num_vertices)]

    # para largura
    self.indice_busca = []
    self.fila = []
    self.nivel = []
    self.diametro = 0
    self.raio = 0
    self.distancia_media = 0

  def get_matrizAdj(self):
    return self.matrizAdj

  def get_numVertices(self):
    return len(self.get_matrizAdj())

  def get_distancia_media(self):
    return self.distancia_media

  def incrementa_tempo(self):
    self.tempo+=1

  def get_tempo(self):
    return self.tempo

  def set_profundidade_entrada(self, vertice, valor):
    self.profundidade_entrada[vertice] = valor

  def get_profundidade_entrada(self, vertice):
    return self.profundidade_entrada[vertice]

  def set_profundidade_saida(self, vertice, valor):
    self.profundidade_saida[vertice] = valor

  def get_profundidade_saida(self, vertice):
    return self.profundidade_saida[vertice]

  def set_pai(self, vertice, valor):
    self.pai[vertice] = valor

  def get_pai(self, vertice):
    return self.pai[vertice]

  def get_vizinhancaAberta(self, vertice):
      vizinhanca_aberta = self.listaAdj[vertice].copy()
      vizinhanca_aberta.pop(0)
      return vizinhanca_aberta

  def pinta_aresta(self, i, j, cor):
    self.matrizCores[i][j] = cor

  def info(self):
    n_vertices = len(self.get_matrizAdj())
    for i in range(n_vertices):
      for j in range(n_vertices):
        print(self.matrizCores[i][j], end = ' ')
      print()

  def inicializa_profundidade(self, vertice_inicial):
    for linha in range(len(self.get_matrizAdj())):
      self.profundidade_entrada.append(0)
      self.profundidade_saida.append(0)
      self.pai.append(0)
    # escolher um vértice aleatório pra iniciar a busca,
    # vamos sempre escolher o primeiro
    self.busca_profundidade(vertice_inicial)

  def busca_profundidade(self, vertice_inicial):
    self.incrementa_tempo()
    self.set_profundidade_entrada(vertice_inicial, self.get_tempo())
    for vizinho in self.get_vizinhancaAberta(vertice_inicial):
      # colocamos o - 1 aqui embaixo pq o vizinho é um vértice e não um índice
      if self.get_profundidade_entrada(vizinho - 1) == 0:
        self.pinta_aresta(vertice_inicial, vizinho - 1, 'azul')
        self.pinta_aresta(vizinho - 1, vertice_inicial, 'azul')
        self.set_pai(vizinho - 1, vertice_inicial)
        self.busca_profundidade(vizinho - 1)

      else:
        # colocamos o - 1 aqui embaixo pq o vizinho é um vértice e não um índice
        if self.get_profundidade_saida(vizinho - 1) == 0 and self.get_pai(vertice_inicial) != vizinho - 1:
          self.pinta_aresta(vertice_inicial, vizinho - 1, 'vermelho')
          self.pinta_aresta(vizinho - 1, vertice_inicial, 'vermelho')

    self.incrementa_tempo()
    self.set_profundidade_saida(vertice_inicial, self.get_tempo())


  # para largura
  def get_indice_busca(self, vertice):
    return self.indice_busca[vertice]

  def set_indice_busca(self, vertice, valor):
    self.indice_busca[vertice] = valor

  def set_nivel(self, vertice, nivel):
    self.nivel[vertice] = nivel

  def get_nivel(self, vertice):
    return self.nivel[vertice]

  def inicializa_largura(self, vertice_inicial):
    self.indice_busca = []
    self.pai = []
    self.nivel = []
    self.fila = []
    self.tempo = 0
    self.raio = 0
    self.diametro = 0
    for linha in range(len(self.get_matrizAdj())):
      self.indice_busca.append(0)
      self.pai.append(0)
      self.nivel.append(0)

    # lembrando que trabalhamos com o índice e não com o vértice em si
    self.set_nivel(vertice_inicial, 0) # como v é uma nova raiz, seu nível é igual a 0
    self.incrementa_tempo()
    self.set_indice_busca(vertice_inicial, self.get_tempo())
    self.fila.append(vertice_inicial)
    self.busca_largura()


  def busca_largura(self):
    # inicializa o loop while enquanto a fila não está vazia
    while self.fila:
      # remove e armazena o 1 vertice da fila para analise. O bfs começa a partir daqui.
      vertice_inicial = self.fila.pop(0)

      # verifica os vertices adjacentes ao inicial escolhido acima, como é aberta ele não pega ele mesmo.
      for vizinho in self.get_vizinhancaAberta(vertice_inicial):

        # verifica se o vizinho ainda não foi visitado, isso é feito com base no índice. O vizinho -1 é justamente pq estamos acessando o vértice e não o índice.
        if self.get_indice_busca(vizinho - 1) == 0:
          # pinta a aresta de forma simétrica entre o vértice inicial e o vizinho de azul, indicando que estão conectados e foram descobertos.
          self.pinta_aresta(vertice_inicial, vizinho - 1, 'azul')
          self.pinta_aresta(vizinho - 1, vertice_inicial, 'azul')
          # atualiza o pai do vertice adjacente como sendo o vértice inicial, que o descobriu
          self.set_pai(vizinho - 1, vertice_inicial)
          # adiciona +1 no nível do vértice que foi descoberto, ou seja o vértice descoberto ganha 1 no seu nível
          self.set_nivel(vizinho - 1, self.get_nivel(vertice_inicial) + 1)
          # incrementa o tempo, como visto nos slides
          self.incrementa_tempo()
          # define o tempo naquele vértice, para retornar o contador de tempo naquele momento, que vai ser incrementado posteriormente
          self.set_indice_busca(vizinho - 1, self.get_tempo())
          # adiciona o vizinho na fila para explorar ele depois
          self.fila.append(vizinho - 1)

        # quando o já foi visitado, ele ainda precisa analisar sua relação com o vertice inciial
        else:
          # analisando se estão no mesmo nível
          if self.get_nivel(vizinho - 1) == self.get_nivel(vertice_inicial):
            # analisando se eles possuem o mesmo pai, se sim pinta de vermelho, vértice descoberto e está sendo revisitado, isso pode indicar um ciclo.
            if self.get_pai(vizinho - 1) == self.get_pai(vertice_inicial):
              self.pinta_aresta(vertice_inicial, vizinho - 1, 'vermelho')
              self.pinta_aresta(vizinho - 1, vertice_inicial, 'vermelho')
            else: # não está no mesmo vértice
              # se eles não tiverem o mesmo pai, pinta de amarelo(cross), ou seja um vértice que não é pai de outro mas estão no mesmo nível.
              self.pinta_aresta(vertice_inicial , vizinho - 1, 'amarelo')
              self.pinta_aresta(vizinho - 1, vertice_inicial, 'amarelo')
          else:
           # se o vertice inicial tiver num nivel acima do adjacente, é um retorno e pinta de verde(um vertice é pai de outro que tá em outra árvore)
            if self.get_nivel(vizinho - 1) == self.get_nivel(vertice_inicial) + 1:
              self.pinta_aresta(vertice_inicial , vizinho - 1, 'verde')
              self.pinta_aresta(vizinho - 1, vertice_inicial, 'verde')


  def gera_excentricidades(self):
    excentridades = []
    self.distancia_media = 0
    # executa a busca em largura(itera) em todos os vértices do grafo
    for vertice in range(self.get_numVertices()):
      # executa a busca em largura naquele vértice, explicada anteriormente
      self.inicializa_largura(vertice)
      print(self.nivel)
      # pega a excentridade rodada por o vértice atual
      print(f'exc({vertice+1}) = {max(self.nivel)}')
      excentridades.append(max(self.nivel))

      # para conseguir a distancia media:
      media = 0
      for nivel in self.nivel:
        media += nivel
      media = media/(len(self.nivel)-1)
      print("Média da busca realizada acima: ", media)
      print("\n")
      self.distancia_media += media

    self.distancia_media/=self.get_numVertices()
    return excentridades

  def get_diametro(self, excentricidades):
    return max(excentricidades)

  def get_raio(self, excentricidades):
    return min(excentricidades)

  def gera_output(self, output_name):
    with open(output_name, 'w') as f:
      f.write('nodedef>name VARCHAR,label VARCHAR\n')
      for vertice in range(self.get_numVertices()):
        f.write(f"{vertice+1},{vertice+1}\n") # passando e nomeando vertice

      f.write('edgedef>node1 VARCHAR,node2 VARCHAR,directed BOOLEAN,color VARCHAR\n')
      for i in range(self.get_numVertices()):
        for j in range(i+1, self.get_numVertices()):
          if self.get_matrizAdj()[i][j]: # se tem aresta
            if self.matrizCores[i][j] == 'vermelho':
              f.write(f"{i+1},{j+1},false,'255,0,0'\n")
            elif self.matrizCores[i][j] == 'verde':
              f.write(f"{i+1},{j+1},false,'0,255,0'\n")
            elif self.matrizCores[i][j] == 'azul':
              f.write(f"{i+1},{j+1},false,'0,0,255'\n")
            elif self.matrizCores[i][j] == 'amarelo':
              f.write(f"{i+1},{j+1},false,'255,255,0'\n")
```

```python
grafo = Grafo('graph_1.txt')
vertice_inicial = 0
grafo.inicializa_profundidade(vertice_inicial)
# grafo.info()
grafo.gera_output('profundidade.gdf')
```

```python
grafo = Grafo('graph_1.txt')
vertice_inicial = 0
grafo.inicializa_largura(vertice_inicial)
# grafo.info()
grafo.gera_output('largura.gdf')
excentricidades = grafo.gera_excentricidades()
print(f'Raio do grafo = {grafo.get_raio(excentricidades)}')
print(f'Diametro do grafo = {grafo.get_diametro(excentricidades)}')
print(f'Distância média do grafo = {grafo.get_distancia_media():.2f}')
```
